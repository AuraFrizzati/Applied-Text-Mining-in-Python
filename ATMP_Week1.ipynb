## **1.2: Regular Expressions**

### **Processing free-text**
In text processing often regular expressions or patterns play a role.

```python
text10 = '"Ethics are built right into the ideals and objectives of the United Nations" \
#UNSG @ NY Society for Ethical Culture bit.ly/2guVelr @UN @UN_Women'
text10
```

```python
## how to extract the hashtag (#) information from this tweet?
text11 = text10.split(' ')
[h for h in text11 if h.startswith('#')]
```

```python
## what about callouts? (they start with '@')
[h for h in text11 if h.startswith('@')]
#the first @ is separated from the words, so it should get excluded
```

```python
text00 = 'blabla aura@google'
[w for w in text00 if re.search('@[A-Za-z0-9_]+', w)]
```

### **Meta-characters: characters matches for regex**
- `.` : wildcard, matches a single character
- `^` : start of a string
- `$` : end of a string
- `[]` : matches one of the set of characters within `[]`
- `[a-z]` : matches one of the range of characters `a, b, ..., z`
- `[^abc]` : matchs=es a character that is **not** `a, b or c`
- `a|b` : matches either `a` or `b`, where `a` and `b` are strings
- `()` : scoping for operators
- `\` : escape character for special characters (`\t`, `\n`, `\b`) 
- `\b` : matches word boundary
- `\d`: any digit, equivalent to `[0-9]`
- `\D` : Any non-digit, equivalent to `[^0-9]`
- `\s` : Any whitespace, equivalent to `[ \t\n\r\f\v]`
- `\S` : Any non-whitespace, equivalent to `[^ \t\n\r\f\v]`
- `\w` : Any alphanumeric character, equivalent to `[a-zA-Z0-9_]`
- `\w` : Any non-alphanumeric character, equivalent to `[^a-zA-Z0-9_]`

**Repetitions**:  
- `*` : matches 0+ occurrences
- `+` : matches 1+ occurrences
- `?` : matches 0 or 1 occurrence
- `{n}` : matches exactly `n` repetitions, with `n` >= o
- `{n,}` : matches at least `n` repetitions, with `n` >= o
- `{,n}` : matches at most `n` repetitions, with `n` >= o
- `{m,n}` : matches at least `m` and at most `n` repetitions, with `m` and `n` >= o

### **Case study: Regular expressions for Dates**
- Date variations for 23rd October 2002:
  - 23-10-2002
  - 23/10/2002
  - 23/10/02
  - 10/23/2002
  - 23 Oct 2002
  - 23 October 2002
  - Oct 23, 2002
  - October 23, 2002

```python
dateStr = '23-10-2002\n23/10/2002\n23/10/02\n10/23/2002\n23 Oct 2002\n23 October 2002\nOct 23, 2002\nOctober 23, 2002\n'
print(dateStr)
```

```python
## to extract only the first 4 dates:
re.findall(r'\d{1,2}[/-]\d{1,2}[/-]\d{2,4}',dateStr) 
# day and month could be 1 or 2 digits
# year could be 2 or 4 digits
```

```python
## to extract only the fifth date:
re.findall(r'\d{1,2} (Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec) \d{2,4}'
,dateStr) 
# only 'Oct' gets extracted because () inside the regex indicate scoping 
# (i.e. selection of a specific pattern from the mathced string)
```

```python
## To extract the whole string, the code needs to be modified to avoid scoping:
re.findall(r'\d{1,2} (?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec) \d{2,4}'
,dateStr)
```

```python
## to extract both the fifth and the sixth date:
re.findall(r'\d{1,2} (?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* \d{2,4}'
,dateStr)
```

```python
## to extract all the last 4 dates:
re.findall(r'(?:\d{1,2} )?(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* (?:\d{1,2}, )?\d{2,4}'
,dateStr)
```

## **1.3: Demonstration: Regex with Pandas and Named Groups**


```python
import pandas as pd

time_sentences = ["Monday: The doctor's appointment is at 2:45pm.", 
                  "Tuesday: The dentist's appointment is at 11:30 am.",
                  "Wednesday: At 7:00pm, there is a basketball game!",
                  "Thursday: Be back home by 11:15 pm at the latest.",
                  "Friday: Take the train at 08:10 am, arrive at 09:00am."]

df = pd.DataFrame(time_sentences, columns=['text'])
df
```

**Regex capturing groups** can be used to group patterns of interest:

```python
df['text'].str.findall(r'(\d?\d):(\d\d)')
```

```python
df['text'].str.replace(r'\w+day\b', '???') ## to replace every weekday with ???
```

```python
df['text'].str.replace(r'(\w+day\b)', lambda x: x.groups()[0][:3]) 
## the weekday is transformed into a group by adding parentheses 
## to regex: r'(\w+day\b)'
```

```python
## to create columns by extracting regex:
df['text'].str.extract(r'(\d?\d):(\d\d)')
```

```python
df['text'].str.extractall(r'((\d?\d):(\d\d) ?([ap]m))')
```

```python
## we can also name groups created with regex:
df['text'].str.extractall(r'(?P<time>(?P<hour>\d?\d):(?P<minute>\d\d)?(?P<period>[ap]m))')
```

## **1.4: Internationalization and Issues with Non-ASCII Characters**

### **English and ASCII**
- English is very well encoded using ASCII
- ASCII = American Standard Code for Information Interchange; it is an encoding schemen that captures all characters:
  - 7-bit character encoding standard: 128 validated codes
  - Exadecimal range: 0x00 - 0x7F [$(0000~0000)_2$ to $(0111~1111)_2$]
  - Includes alphabets (upper and lower cases), digits, punctuations, common symbols (e.g. brackets, %, $, #, ...), control characters (e.g. newline character or tab)
  - Worked (relatively) well for English typewriting, although it does not really capture everything: diacritics, international languages:

#### **UTF-8**
- Unicode Transformational Format - 8 bits
- Variable length encoding: 1 to 4 bytes
